# Assigment
Write a function that adds any two matrices [A], [B] (compatible by sum). Both [A], [B] must belong to the same linear space $R_n^m$

```python
A = [ [1,2,3], [4,5,6], [7,8,9] ]
B = [ [10,20,30], [40,50,60], [70,80,90] ]
```

```python
def MATSUM(args):
    return  AA(AA(SUM)) (AA(TRANS)(TRANS(args)))
```

```python
MATSUM([A,B])
```

Vector product w of two $R^3$ vectors

```python
def VECTPROD(args):
    u,v = args
    w = N(3)(0)

    w[0] = u[1]*v[2] - u[2]*v[1]
    w[1] = u[2]*v[0] - u[0]*v[2]
    w[2] = u[0]*v[1] - u[1]*v[0]
    return w

vp = VECTPROD([[1,0,0], [0,1,0]])
vp
```

# Random points

```python
from random import random

def randomPoints(m, sx=1, sy=1):
    def point():
        return [random() * sx, random() * sy]
    return [point() for k in range(m)]
```

```python
verts = randomPoints(200, 2*PI, 2)
verts
```

```python
obj = MKPOL([verts, AA(LIST)(range(200)), None])
VIEW(obj)
```
